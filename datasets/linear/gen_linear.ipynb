```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from gen_linear import create_poly, create_regression, create_classification
from sklearn.model_selection import train_test_split
```

```python
x0_1, x1_1 = create_poly(seed=1234, n=1000, noise=0.5, bias=2.5)
x0_2, x1_2 = create_poly(seed=1234, n=1000, noise=0.5, bias=-2.5)

fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(16,10))
ax.scatter(x0_1[:,1],x1_1)
ax.scatter(x0_2[:,1],x1_2)
ax.set_title('Linear Data using PolynomialFeatures')
plt.show()
```

```python
y_1 = np.zeros(x0_1[:,1].shape)
_dataset = np.zeros((x0_1.shape[0], 3))
_dataset[:,0:] = x0_1[:,1:]
_dataset[:,1:] = x1_1
_dataset[:,2] = y_1
dataset = pd.DataFrame(_dataset, columns=['x0','x1','label'])

y_2 = np.ones(x0_2[:,1].shape)
_dataset = np.zeros((x0_2.shape[0], 3))
_dataset[:,0:] = x0_2[:,1:]
_dataset[:,1:] = x1_2
_dataset[:,2] = y_2
dataset = pd.concat([dataset, pd.DataFrame(_dataset, columns=['x0','x1','label'])]).reset_index(drop=True)
```

```python
x0_1, x1_1 = create_regression(bias=120)
x0_2, x1_2 = create_regression(bias=-120)

fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(16,10))
ax.scatter(x0_1[:,1],x1_1)
ax.scatter(x0_2[:,1],x1_2)
ax.set_title('Linear Data using make_regression')
plt.show()
```

```python
X, y = create_classification()
```

```python
X, y = pd.DataFrame(X, columns=['x0', 'x1']), pd.DataFrame(y, columns=['label'])
dataset=pd.concat([X, y], axis=1)
```

```python
#split the data into train and test set
train, test = train_test_split(dataset, test_size=0.2, random_state=0)
#save the data
train.to_csv('./train.csv',index=False)
test.to_csv('./test.csv',index=False)
```
