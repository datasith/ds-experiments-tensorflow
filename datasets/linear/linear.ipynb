```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# Reference:
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html
# It is very easy to make the dataset more or less challenging w flip_y and class_sep. 
# Setting n_informative and class_sep to higher values will make it less challenging.

rnd = 1234
X, y = make_classification(n_samples=5000, n_features=2, n_redundant=0,
      n_informative=2, n_classes=2, random_state=rnd,
      n_clusters_per_class=1, class_sep=3, hypercube=True)

X, y = pd.DataFrame(X, columns=['x0', 'x1']), pd.DataFrame(y, columns=['label'])
dataset=pd.concat([X, y], axis=1)
```

```python
plt.title("Two informative features, one cluster per class", fontsize="small")
_ = plt.scatter(X['x0'], X['x1'], marker="o", c=y['label'], s=25, edgecolor="k")
```

```python
#split the data into train and test set
train, test = train_test_split(dataset, test_size=0.2, random_state=0)
#save the data
train.to_csv('./train.csv',index=False)
test.to_csv('./test.csv',index=False)
```
